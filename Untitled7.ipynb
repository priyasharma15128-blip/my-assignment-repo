```python
# ==============================
# Student Performance Dashboard
# Using Dash (Python)
# ==============================

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# -----------------------------
# 1. Create Sample Dataset
# -----------------------------
data = {
    "Student_ID": ["S001", "S002", "S003", "S004", "S005", "S006", "S007"],
    "Name": ["Asha", "Ravi", "Neha", "Amit", "Kiran", "Priya", "Rahul"],
    "Gender": ["F", "M", "F", "M", "F", "F", "M"],
    "Class": [10, 10, 10, 9, 9, 10, 9],
    "Math": [85, 56, 95, 70, 82, 78, 65],
    "Science": [78, 60, 92, 68, 88, 85, 70],
    "English": [88, 65, 89, 75, 80, 90, 72],
    "Attendance": [92, 85, 97, 90, 95, 93, 88]
}

df = pd.DataFrame(data)
df["Overall_Percentage"] = df[["Math", "Science", "English"]].mean(axis=1)

# -----------------------------
# 2. Initialize Dash App
# -----------------------------
app = dash.Dash(__name__)

# -----------------------------
# 3. Dashboard Layout
# -----------------------------
app.layout = html.Div([
    html.H1("ðŸŽ“ Student Performance Dashboard",
            style={'textAlign': 'center', 'color': '#003366'}),

    # Filters
    html.Div([
        html.Div([
            html.Label("Select Class:"),
            dcc.Dropdown(
                id='class-dropdown',
                options=[{'label': str(c), 'value': c} for c in sorted(df['Class'].unique())],
                value=sorted(df['Class'].unique())[0],
                clearable=False
            )
        ], style={'width': '30%', 'display': 'inline-block', 'padding': '10px'}),

        html.Div([
            html.Label("Select Gender:"),
            dcc.Dropdown(
                id='gender-dropdown',
                options=[{'label': g, 'value': g} for g in df['Gender'].unique()] + [{'label': 'All', 'value': 'All'}],
                value='All',
                clearable=False
            )
        ], style={'width': '30%', 'display': 'inline-block', 'padding': '10px'})
    ]),

    # Graphs
    dcc.Graph(id='subject-bar'),
    dcc.Graph(id='attendance-scatter'),
    dcc.Graph(id='gender-pie'),

    html.H3("ðŸ“‹ Student Details", style={'textAlign': 'center'}),
    html.Div(id='data-table')
])

# -----------------------------
# 4. Callbacks for Interactivity
# -----------------------------
@app.callback(
    [Output('subject-bar', 'figure'),
     Output('attendance-scatter', 'figure'),
     Output('gender-pie', 'figure'),
     Output('data-table', 'children')],
    [Input('class-dropdown', 'value'),
     Input('gender-dropdown', 'value')]
)
def update_dashboard(selected_class, selected_gender):
    # Filter data
    filtered_df = df[df['Class'] == selected_class]
    if selected_gender != 'All':
        filtered_df = filtered_df[filtered_df['Gender'] == selected_gender]

    # Bar Chart: Subject-wise marks
    melted = filtered_df.melt(id_vars=['Name'],
                              value_vars=['Math', 'Science', 'English'],
                              var_name='Subject', value_name='Marks')
    bar_fig = px.bar(melted, x='Subject', y='Marks', color='Name',
                     title=f"Subject-wise Marks (Class {selected_class})")

    # Scatter Plot: Attendance vs Performance
    scatter_fig = px.scatter(filtered_df, x='Attendance', y='Overall_Percentage',
                             color='Name', size='Overall_Percentage',
                             title="Attendance vs Overall Performance", trendline='ols')

    # Pie Chart: Gender distribution
    gender_fig = px.pie(filtered_df, names='Gender', title='Gender Distribution')

    # Data Table
    table_header = [html.Thead(html.Tr([html.Th(col) for col in filtered_df.columns]))]
    table_body = [html.Tbody([
        html.Tr([html.Td(filtered_df.iloc[i][col]) for col in filtered_df.columns])
        for i in range(len(filtered_df))
    ])]
    table = html.Table(table_header + table_body,
                       style={'width': '100%', 'border': '1px solid #ccc', 'textAlign': 'center'})

    return bar_fig, scatter_fig, gender_fig, table

# -----------------------------
# 5. Run the Server
# -----------------------------
if __name__ == '__main__':
    app.run(debug=True)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# -----------------------------
# 1. Create Sample Sales Dataset
# -----------------------------
sales_data = {
    "Product": ["A", "B", "C", "A", "B", "C", "A", "B", "C", "D", "E", "D", "E", "D", "E"],
    "Region": ["North", "South", "East", "North", "South", "East", "North", "South", "East", "North", "South", "East", "North", "South", "East"],
    "Sales": [150, 200, 100, 180, 220, 120, 160, 210, 110, 130, 170, 150, 190, 140, 180],
    "Month": ["Jan", "Jan", "Jan", "Feb", "Feb", "Feb", "Mar", "Mar", "Mar", "Jan", "Jan", "Feb", "Feb", "Mar", "Mar"]
}

sales_df = pd.DataFrame(sales_data)

# -----------------------------
# 2. Initialize Dash App
# -----------------------------
app_sales = dash.Dash(__name__)

# -----------------------------
# 3. Sales Dashboard Layout
# -----------------------------
app_sales.layout = html.Div([
    html.H1("ðŸ“Š Sales Dashboard",
            style={'textAlign': 'center', 'color': '#0056b3'}),

    # Filters
    html.Div([
        html.Div([
            html.Label("Select Region:"),
            dcc.Dropdown(
                id='region-dropdown',
                options=[{'label': r, 'value': r} for r in sales_df['Region'].unique()] + [{'label': 'All', 'value': 'All'}],
                value='All',
                clearable=False
            )
        ], style={'width': '30%', 'display': 'inline-block', 'padding': '10px'}),

        html.Div([
            html.Label("Select Product:"),
            dcc.Dropdown(
                id='product-dropdown',
                options=[{'label': p, 'value': p} for p in sales_df['Product'].unique()] + [{'label': 'All', 'value': 'All'}],
                value='All',
                clearable=False
            )
        ], style={'width': '30%', 'display': 'inline-block', 'padding': '10px'})
    ]),

    # Graphs
    dcc.Graph(id='sales-bar'),
    dcc.Graph(id='sales-line')
])

# -----------------------------
# 4. Callbacks for Interactivity
# -----------------------------
@app_sales.callback(
    [Output('sales-bar', 'figure'),
     Output('sales-line', 'figure')],
    [Input('region-dropdown', 'value'),
     Input('product-dropdown', 'value')]
)
def update_sales_dashboard(selected_region, selected_product):
    # Filter data
    filtered_sales_df = sales_df.copy()
    if selected_region != 'All':
        filtered_sales_df = filtered_sales_df[filtered_sales_df['Region'] == selected_region]
    if selected_product != 'All':
        filtered_sales_df = filtered_sales_df[filtered_sales_df['Product'] == selected_product]

    # Bar Chart: Sales by Product
    bar_fig = px.bar(filtered_sales_df, x='Product', y='Sales', color='Region',
                     title="Sales by Product and Region")

    # Line Chart: Sales over Time (Month)
    line_fig = px.line(filtered_sales_df.groupby('Month')['Sales'].sum().reset_index(),
                       x='Month', y='Sales', title="Total Sales Over Time")


    return bar_fig, line_fig

# -----------------------------
# 5. Run the Server
# -----------------------------
if __name__ == '__main__':
    app_sales.run(debug=True, port=8051)
```

# Task

## Data acquisition

### Subtask:
Find and load a suitable dataset for COVID-19 data. This might involve using a library or downloading a file.


**Reasoning**:
Search for a suitable COVID-19 dataset, download it, and load it into a pandas DataFrame. A widely used dataset is from Johns Hopkins University, available on GitHub.



```python
import requests
import pandas as pd

# URL for the daily reports from Johns Hopkins University CSSEGISandData
# Using a specific date for demonstration; in a real application, you might automate this.
# As of late 2023, the daily reports are structured differently,
# but the time series data is still a good source.
# Let's use the time series confirmed cases global data.
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"

try:
    # Download the data
    response = requests.get(url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

    # Load the data into a pandas DataFrame
    from io import StringIO
    df_covid = pd.read_csv(StringIO(response.text))

    print("Dataset loaded successfully.")
    print(df_covid.head())

except requests.exceptions.RequestException as e:
    print(f"Error downloading the dataset: {e}")
    df_covid = None
except pd.errors.EmptyDataError:
    print("The downloaded file is empty.")
    df_covid = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df_covid = None
```

## Data preprocessing

### Subtask:
Clean and prepare the data for analysis and visualization. This could include handling missing values, converting data types, and aggregating data as needed.


**Reasoning**:
Melt the DataFrame, convert the 'Date' column to datetime objects, handle missing values by filling them with 0, aggregate the data by 'Country/Region' and 'Date', and calculate the daily new cases for each country.



```python
# Melt the DataFrame to transform date columns
df_melted = df_covid.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Confirmed'
)

# Convert 'Date' column to datetime objects
df_melted['Date'] = pd.to_datetime(df_melted['Date'], errors='coerce')

# Handle missing values in 'Confirmed' by filling with 0
df_melted['Confirmed'] = df_melted['Confirmed'].fillna(0).astype(int)

# Aggregate the data by 'Country/Region' and 'Date'
df_agg = df_melted.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()

# Calculate daily new cases for each country
df_agg['New_Cases'] = df_agg.groupby('Country/Region')['Confirmed'].diff().fillna(0).astype(int)

# Ensure new cases are not negative (can happen due to data revisions)
df_agg['New_Cases'] = df_agg['New_Cases'].apply(lambda x: max(0, x))

# Display the first few rows of the processed data
display(df_agg.head())
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Initialize the Dash application
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("COVID-19 Confirmed Cases Dashboard",
            style={'textAlign': 'center', 'color': '#1f77b4'}),

    # Filters section
    html.Div([
        html.Label("Select Country/Region:"),
        dcc.Dropdown(
            id='country-dropdown',
            options=[{'label': country, 'value': country} for country in sorted(df_agg['Country/Region'].unique())],
            value='World', # Default value
            clearable=False
        )
    ], style={'width': '50%', 'margin': 'auto', 'padding': '10px'}),

    # Graphs section
    dcc.Graph(id='confirmed-cases-line-chart'),
    dcc.Graph(id='new-cases-bar-chart')
])

# Note: Callbacks to update the graphs based on the dropdown selection will be defined in the next subtask.

# To run the app, you would typically include:
# if __name__ == '__main__':
#     app.run_server(debug=True)
```

```python
@app.callback(
    [Output('confirmed-cases-line-chart', 'figure'),
     Output('new-cases-bar-chart', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graph(selected_country):
    if selected_country == 'World':
        filtered_df = df_agg.groupby('Date')['Confirmed', 'New_Cases'].sum().reset_index()
        title_suffix = "Worldwide"
    else:
        filtered_df = df_agg[df_agg['Country/Region'] == selected_country].copy()
        title_suffix = f"in {selected_country}"

    # Create line chart for confirmed cases
    line_fig = px.line(filtered_df, x='Date', y='Confirmed',
                       title=f'Total Confirmed Cases {title_suffix}')

    # Create bar chart for new cases
    bar_fig = px.bar(filtered_df, x='Date', y='New_Cases',
                     title=f'Daily New Cases {title_suffix}')

    return line_fig, bar_fig

# To run the app, you would typically include:
# if __name__ == '__main__':
#     app.run_server(debug=True)
```
